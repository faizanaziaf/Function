**Q1: What is the difference between a function and a method in Python?**  
**A1:** A function is a block of code that performs an action and is defined outside of classes, whereas a method is a function associated with an object and defined within a class.  
*Example:* `len()` is a function, but `str.upper()` is a method.

**Q2: Explain the concept of function arguments and parameters in Python.**  
**A2:** Parameters are the placeholders defined in a function, while arguments are the actual values passed to the function when it is called.  
*Example:* In `def greet(name):`, `name` is a parameter. In `greet("Alice")`, `"Alice"` is an argument.

**Q3: What are the different ways to define and call a function in Python?**  
**A3:** Functions can be defined using `def` or `lambda`. They are called by using their name followed by parentheses with arguments (if any).  
*Example:* Using `def`: `def add(a, b): return a + b`; call it as `add(2, 3)`.  
Using `lambda`: `add = lambda a, b: a + b`; call it as `add(2, 3)`.

**Q4: What is the purpose of the `return` statement in a Python function?**  
**A4:** The `return` statement sends the result of the function's computation back to the caller.  
*Example:* `def square(x): return x**2` will return `4` for `square(2)`.

**Q5: What are iterators in Python and how do they differ from iterables?**  
**A5:** An iterator is an object with a `__next__()` method to fetch the next value, while an iterable is an object that can be looped over but needs an iterator to retrieve values.  
*Example:* A list (`[1, 2, 3]`) is iterable. Using `iter([1, 2, 3])` creates an iterator.

**Q6: Explain the concept of generators in Python and how they are defined.**  
**A6:** Generators are functions that yield values one at a time using the `yield` keyword, making them memory-efficient.  
*Example:* `def my_gen(): yield 1; yield 2; yield 3` creates a generator.

**Q7: What are the advantages of using generators over regular functions?**  
**A7:** Generators are memory-efficient, as they generate values lazily and don't store them in memory.  
*Example:* `range(10**6)` is a generator, whereas `[i for i in range(10**6)]` creates a list in memory.

**Q8: What is a lambda function in Python and when is it typically used?**  
**A8:** A lambda function is a short, anonymous function defined with `lambda` and used for simple operations.  
*Example:* `lambda x: x**2` can be used in `map(lambda x: x**2, [1, 2, 3])`.

**Q9: Explain the purpose and usage of the `map()` function in Python.**  
**A9:** The `map()` function applies a given function to each item of an iterable and returns an iterator.  
*Example:* `map(str.upper, ["a", "b", "c"])` returns `["A", "B", "C"]`.

**Q10: What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**  
**A10:** `map()` transforms all items in an iterable, `filter()` selects items based on a condition, and `reduce()` aggregates items into a single result.  
*Example:* `map(lambda x: x**2, [1, 2]) → [1, 4]`; `filter(lambda x: x > 1, [1, 2]) → [2]`; `reduce(lambda x, y: x + y, [1, 2]) → 3`.
